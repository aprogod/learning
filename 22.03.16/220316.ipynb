```python
sql_13 = """
            SELECT
            DEPTNO,
            GROUP_CONCAT(ENAME),
            GROUP_CONCAT(JOB)
            FROM emp
            GROUP BY DEPTNO
         """

### CONCAT : 연결하는 함수

result_13 = _dt._executeAll(sql_13)
df_13 = pd.DataFrame(result_13)
df_13
```

```python
sql_16 = """
            SELECT ENAME AS "이름",
            CASE
               WHEN JOB = "ANALYST" THEN "분석가"
               WHEN JOB = "CLERK" THEN "사원"
               WHEN JOB = "MANAGER" THEN "관리자"
               WHEN JOB = "PRESIDENT" THEN "사장"
               ELSE "판매원"
               END AS "직업"
            FROM emp
         """

result_16 = _dt._executeAll(sql_16)
df_16 = pd.DataFrame(result_16)
df_16
```
