1. emp table 과 dept table을 left join 조건 dept가 같은 경우
2. 출력 컬럼은 ENAME, JOB SAL, DEPTNO, DNAME, LOC
3. ENAME은 그대로 출력
4. JOB은 각각 한글명으로 변경
5. SAL은 2450을 기준으로 2450 이상이면 HIGH 미만이면 LOW
6. DEPTNO은 DEPTNO의 값과 dept table의 DNAME 값을 더해서 표시 (DEPTNO-DNAME 형식)
7. 모든 컬럼의 이름은 한글로 변경

```python
sql_17 = """
            SELECT
            
            emp.ENAME AS 이름,
            IF(emp.SAL >= 2450, 'HIGH', 'LOW') AS 판매,
            CASE
               WHEN emp.JOB = "ANALYST" THEN "분석가"
               WHEN emp.JOB = "CLERK" THEN "사원"
               WHEN emp.JOB = "MANAGER" THEN "관리자"
               WHEN emp.JOB = "PRESIDENT" THEN "사장"
               ELSE "판매원"
               END AS 직업,
            CONCAT(emp.DEPTNO, '-', dept.DNAME) AS 연결,
            dept.LOC AS 위치
            
            FROM emp
            LEFT JOIN dept
            ON emp.DEPTNO = dept.DEPTNO
            
         """

result_17 = _dt._executeAll(sql_17)
df_17 = pd.DataFrame(result_17)
df_17
```
