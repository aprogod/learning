```python
import mod_sql
```

```python
_db = mod_sql.Database()
sql = "SELECT *FROM btc"
result = _db._executeAll(sql)
dt = pd.DataFrame(result)
dt
```

```python
dt_1 = dt
dt_1.columns = ["날짜", "시가", "고가", "저가", "종가", "거래량", "가치"]
```

```python
dt_1
```

```python
dt_2 = dt_1
dt_2["거래량"] = dt_2["거래량"].round(2)
dt_2
```

```python
dt_2["등락폭"] = dt_2["시가"]-dt_2["종가"]
dt_2
```

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.plot(dt_2["시가"],'r')
plt.title("open")

plt.subplot(2,2,2)
plt.plot(dt_2["종가"],'g')
plt.title("close")

plt.subplot(2,2,3)
plt.plot(dt_2["등락폭"],'y')
plt.title("diff")

plt.subplot(2,2,4)
plt.plot(dt_2["거래량"],'k')
plt.title("volume")


plt.show()
```
