```python
x = np.linspace(-5,5,101)
y = (1 / np.sqrt(2 *np.pi))*np.exp(-x**2/2)
y
```

```python
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.show
```

```python
import scipy.stats as stats
```

```python
y_ = stats.norm(0,1).pdf(x)
y_
```

```python
plt.plot(x,y_)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
def norm(a,b,x):
    return stats.norm(a,b).pdf(x)

x= np.linspace(-20,20, 101)
x
```

```python
for i in range(1, 11, 1):
    _y = norm(0, i, x)
    plt.plot(x,_y)

plt.show()
```

```python
x = np.random.randint(10,50,5)
labels = ["a", "b", "c", "d", "e"]
plt.pie(x, autopct="%.1f%%", labels=labels)
plt.show()
```

```python
explodes = [0, 0.5, 0.5, 0, 0]
plt.pie(x, autopct="%.1f%%", labels=labels, counterclock=True, explode = explodes)
plt.show()
```
