### 1번 문제
> For문을 이용하여 구구단을 출력하라
    2단부터 9단까지 구구단을 print를 이용하여 출력하라.

```python
for i in range(2,10,1):
    for j in range(2,10,1):
        print("%d x %d = %d"%(i,j,(i*j)))
```

### 2번 문제
> 2개의 주사위를 던져서 두 수의 합이 5의 배수인 경우를 출력하라.
1. 2개의 for문을 이용할 것
2. If문을 이용하여 5의 배수 조건을 줄 것

```python
for i in range(1,7):
    for j in range(1,7):
        if (i +j)%5 == 0:
            print("%d + %d = %d"%(i,j,i+j))
```

### 3번 문제
> 3개의 숫자를 입력하면 (input_a * input_b + input_c)와 (input_a + input_b * input_c)의 차이를 나타내는 함수를 만드시오.
1. 함수명은 diff_sum()이라 지정
2. 함수에 매개변수를 3개 지정(input_a, input_b, input_c)
3. (input_a * input_b + input_c)와 (input_a + input_b * input_c) 두 식의 결과 값은 언제나 큰 수에서 작은 수를 빼는 if 조건문 생성
4. 3개의 숫자는 input를 사용하여 입력 값을 받을 것

```python
def diff_sum(input_a, input_b, input_c):
    x = input_a * input_b + input_c
    y = input_a + input_b * input_c
    if x > y :
        print(x-y)
    elif x < y:
        print(y-x)
    else :
        print("두 값이 같습니다.")
```

### 4번 문제
> Pandas를 이용하여 다음과 같은 데이터프레임을 만들고 데이터프레임을 수정하라.

|   | 0 | 1 |  2 |
|:-:|:-:|:-:|:--:|
| 0 | 1 | 2 |  3 |
| 1 | 7 | 8 |  9 |
| 2 | 4 | 5 |  6 |
| 3 | 3 | 5 | 10 |

1. 컬럼의 이름을 A, B, C로 변경
2. D라는 새로운 컬럼에 shift 함수를 이용하여 C 컬럼의 차를 D 컬럼에 생성(NaN값인 경우에는 0으로 대체)
3. A 컬럼을 기준으로 오름차순 정렬
4. 인덱스의 값을 새로 지정(기존의 인덱스는 삭제)
5. 이 모든 과정은 원본의 데이터를 수정 

```python
import pandas as pd
list_1 = [[1,2,3],[7,8,9],[4,5,6],[3,5,10]]
df = pd.DataFrame(list_1)
df
```

```python
# 1. 컬럼의 이름을 A, B, C로 변경
df.columns=["A", "B", "C"]
df
```

```python
# 2. D라는 새로운 컬럼에 shift 함수를 이용하여 C 컬럼의 차를 D 컬럼에 생성(NaN값인 경우에는 0으로 대체)
df_1 = df ### 원본 살려두기
df_1["D"] = (df_1["C"]-df_1["C"].shift()).fillna(0)
df_1
```

```python
# 3. A 컬럼을 기준으로 오름차순 정렬
df_1.sort_values("A", inplace=True)
df_1
```

```python
# 4. 인덱스의 값을 새로 지정(기존의 인덱스는 삭제)
df_1.reset_index(drop=True, inplace=True)
df_1
```

### 5번 문제
> csv파일을 읽어온 뒤 데이터프레임을 수정하는 Class를 만들어보자.

1. class를 생성할 때, 매개 변수를 이용하여 csv 파일 주소를 받아 온 후 csv파일 읽기(파일 주소는 상대경로로 지정)
2. class 내의 함수 _sort()를 생성 후 컬럼의 이름이나 컬럼의 위치(예 : n번째 컬럼)를 받아오는 매개 변수를 생성. 그 값을 기준으로 sort를 하는 함수 생성(매개 변수가 위치(int)인지 이름(str)인지 체크하여 함수 구성, 오름차순 정렬)
3. class 내의 함수 _drop()을 생성하여 매개 변수를 이용하여 컬럼의 이름을 받아온 후 해당 컬럼을 삭제하는 함수 생성
4. class 내의 함수 _reset()을 생성하여 인덱스의 값을 초기화 하고 기존의 인덱스 값은 삭제하는 함수 생성
5. 모든 과정에서 원본의 데이터를 수정

```python

class Class_1():

    # 1. class를 생성할 때, 매개 변수를 이용하여 csv 파일 주소를 받아 온 후 csv파일 읽기(파일 주소는 상대경로로 지정)
    def __init__(self, input_url):
        self.sales_df = pd.read_csv(input_url)
    
    # 2. class 내의 함수 _sort()를 생성 후 컬럼의 이름이나 컬럼의 위치(예 : n번째 컬럼)를 받아오는 매개 변수를 생성.
    # 그 값을 기준으로 sort를 하는 함수 생성(매개 변수가 위치(int)인지 이름(str)인지 체크하여 함수 구성, 오름차순 정렬)
    def _sort(self, input, input_inplace = True):
        self.sales_df.sort_values(input, inplace = input_inplace)
        return self.sales_df
    
    # 3. class 내의 함수 _drop()을 생성하여 매개 변수를 이용하여 컬럼의 이름을 받아온 후 해당 컬럼을 삭제하는 함수 생성
    def _drop(self,*input_columns,input_inplace = True):
        self.input = list(set(input_columns))
        self.sales_df.drop(self.input, axis = "columns", inplace = input_inplace)
        return self.sales_df

    # 4. class 내의 함수 _reset()을 생성하여 인덱스의 값을 초기화 하고 기존의 인덱스 값은 삭제하는 함수 생성
    def _reset(self,input_inplace = True):
        self.sales_df.reset_index(drop=True, inplace = input_inplace)
        return self.sales_df
```

```python
# 1번 실행
test = Class_1("./Sales Records.csv")
```

```python
# 2번 실행
test._sort("Country").head(5)
```

```python
# 3번 실행
test._drop("Ship Date", "Order ID", "Sales Channel", "Total Revenue").head(5)
```

```python
# 4번 실행
test._reset().head(5)
```
